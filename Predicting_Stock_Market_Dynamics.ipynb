# Data Visualizations

```python
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


df = pd.read_csv('Stock_Market_Data.csv')

df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df.set_index('Date', inplace=True)
yearly_data = df.resample('Y').mean()  
yearly_data.to_csv('yearly_stock_data.csv')
plt.figure(figsize=(10, 6))
plt.bar(yearly_data.index.year, yearly_data['Volume'], color='skyblue')
plt.title('Average Volume Traded Per Year')
plt.xlabel('Year')
plt.ylabel('Average Volume')
plt.xticks(yearly_data.index.year, rotation=45) 

plt.tight_layout() 
plt.savefig("100.png")
plt.show()
```

# ML model creation

## LOAD DATA

```python
import pandas as pd
df=pd.read_csv("Stock_Market_Data.csv")
```

## DATA PREPARATION

```python
df
```

### DATA SEPERATION AS X AND Y

```python
df.isnull().sum()
```

```python
features = ['Low', 'Open', 'Volume', 'High']
target = 'Close'
x=df[features]
y=df[target]
```

### DATA SPLITTING

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
x_train
```

```python
x_test
```

### LINEAR REGRESSION

#### **TRAINING THE MODEL**

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

####**APPLYING THE MODEL TO MAKE A PREDICTION**

```python
y_lr_train_pred=lr.predict(x_train)
y_lr_test_pred=lr.predict(x_test)
```

```python
len(y_lr_train_pred)
```

```python
y_lr_test_pred
```

#### **EVALUTING MODEL PERFORMANCE**

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

lr_train_mse=mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2=r2_score(y_train,y_lr_train_pred)
lr_train_mae = mean_absolute_error(y_test, y_lr_test_pred)

lr_test_mse=mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2=r2_score(y_test,y_lr_test_pred)
lr_test_mae = mean_absolute_error(y_test, y_lr_test_pred)
```

```python
lr_train_mse
```

```python
lr_train_r2
```

```python
print("LR MSE (TRAIN):",lr_train_mse)
print("LR R2 (TRAIN):",lr_train_r2)
print("LR MSE (TEST):",lr_test_mse)
print("LR R2 (TEST):",lr_test_r2)
```

```python
lr_results=pd.DataFrame(["Linear Regression",lr_train_mse,lr_train_r2,lr_train_mae,lr_test_mse,lr_test_r2,lr_test_mae]).transpose()
lr_results.columns=['METHOD','TRAINING MSE','TRAINING R2','TRAINING MAE','TEST MSE','TEST R2','TEST MAE']
```

```python
lr_results
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.regplot(x=y_test, y=y_lr_test_pred, line_kws={"color": "red"})
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.savefig('regplot.png')
plt.show()
```

### RANDOM FOREST

#### **TRAINING THE MODEL**

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)
```

#### **APPLYING THE MODEL TO MAKE A PREDICTION**

```python
y_rf_train_pred=rf.predict(x_train)
y_rf_test_pred=rf.predict(x_test)
```

```python
y_rf_test_pred
```

#### **EVALUTING MODEL PERFORMANCE**

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

rf_train_mse=mean_squared_error(y_train,y_rf_train_pred)
rf_train_r2=r2_score(y_train,y_rf_train_pred)
rf_train_mae = mean_absolute_error(y_test, y_rf_test_pred)

rf_test_mse=mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2=r2_score(y_test,y_rf_test_pred)
rf_test_mae = mean_absolute_error(y_test, y_rf_test_pred)
```

```python
rf_results=pd.DataFrame(["Random Forest Regression",rf_train_mse,rf_train_r2,rf_test_mae,rf_test_mse,rf_test_r2,rf_test_mae]).transpose()
rf_results.columns=['METHOD','TRAINING MSE','TRAINING R2','TRAINING MAE','TEST MSE','TEST R2','TEST MAE']
rf_results
```

### MODEL COMPARISION

```python
df_models=pd.concat([lr_results,rf_results],axis=0).reset_index(drop=True)
df_models
```

```python
y_test
```

```python
y_rf_train_pred
```

```python
import matplotlib.pyplot as plt

sns.regplot(x=y_test, y=y_rf_test_pred, data=rf_results, line_kws={"color": "red"})
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted with Regression Line')
plt.savefig('rf.png')
plt.show()


```

# Support Vector Machines (SVMs)


```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
svm = SVR(kernel='rbf', C=100, gamma=0.1)
svm.fit(X_train_scaled, y_train)

y_svm_test_pred = svm.predict(X_test_scaled)
y_svm_train_pred = svm.predict(X_train_scaled)
```

```python
y_svm_test_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

tr_mse_svm = mean_squared_error(y_train, y_svm_train_pred)
tr_mae_svm = mean_absolute_error(y_train, y_svm_train_pred)
tr_r2_svm = r2_score(y_train, y_svm_train_pred)

ts_mse_svm = mean_squared_error(y_test, y_svm_test_pred)
ts_mae_svm = mean_absolute_error(y_test, y_svm_test_pred)
ts_r2_svm = r2_score(y_test, y_svm_test_pred)
```

```python
import pandas as pd
svm_results = pd.DataFrame(["Support Vector Machine", tr_mse_svm, tr_mae_svm, tr_r2_svm,ts_mse_svm, ts_mae_svm, ts_r2_svm]).transpose()
svm_results.columns = ['METHOD','TRAINING MSE','TRAINING R2','TRAINING MAE','TEST MSE','TEST R2','TEST MAE']


```

```python
svm_results
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.regplot(x=y_train, y=y_svm_train_pred, data=svm_results, line_kws={"color": "red"})
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted with Regression Line')
plt.savefig('svm_regplot.png')
plt.show()
```

## K-Nearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train_scaled, y_train)

y_knn_test_pred = knn.predict(X_test_scaled)
y_knn_train_pred = knn.predict(X_train_scaled)
```

```python
y_knn_test_pred
```

```python
tr_mse_knn = mean_squared_error(y_train, y_knn_train_pred)
tr_mae_knn = mean_absolute_error(y_train, y_knn_train_pred)
tr_r2_knn = r2_score(y_train, y_knn_train_pred)

ts_mse_knn = mean_squared_error(y_test, y_knn_test_pred)
ts_mae_knn = mean_absolute_error(y_test, y_knn_test_pred)
ts_r2_knn = r2_score(y_test, y_knn_test_pred)
```

```python
knn_results = pd.DataFrame(["KNN", tr_mse_knn, tr_mae_knn, tr_r2_knn,ts_mse_knn, ts_mae_knn, ts_r2_knn]).transpose()
knn_results.columns = ['METHOD','TRAINING MSE','TRAINING R2','TRAINING MAE','TEST MSE','TEST R2','TEST MAE']
print("KNN Model Evaluation Metrics:")
knn_results
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.regplot(x=y_train, y=y_knn_train_pred, data=knn_results, line_kws={"color": "red"})
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted with Regression Line')
plt.savefig('knn.png')
plt.show()
```

## MODEL COMPARISION

```python
df_models=pd.concat([lr_results,svm_results,rf_results,knn_results],axis=0).reset_index(drop=True)
df_models
```

```python
best_model_test_mse = df_models.loc[df_models['TEST MSE'].astype(float).idxmin()]['METHOD']
best_model_test_mae = df_models.loc[df_models['TEST MAE'].astype(float).idxmin()]['METHOD']
best_model_test_r2 = df_models.loc[df_models['TEST R2'].astype(float).idxmax()]['METHOD']


print(f"The best model based on TEST MSE is: {best_model_test_mse}")
print(f"The best model based on TEST MAE is: {best_model_test_mae}")
print(f"The best model based on TEST R2 is: {best_model_test_r2}")
```
